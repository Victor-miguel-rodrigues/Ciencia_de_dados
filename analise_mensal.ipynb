```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


arquivo = pd.read_csv('mes.csv')

arquivo.info()
```

```python
arquivo['data'] = pd.to_datetime(arquivo['data'], format='%Y-%m-%d')
arquivo['quantidade'] = pd.to_numeric(arquivo['quantidade'].astype(int), errors='coerce')
arquivo['quantidade'] = arquivo['quantidade'].fillna(0)
arquivo['preco_unitario'] = pd.to_numeric(arquivo['preco_unitario'].astype(float))

display(arquivo.info())
```

Você TEM que descobrir:

qual produto mais vende?

qual categoria gera mais dinheiro?

qual mês teve mais vendas?

qual vendedor vende mais?

```python
#display(arquivo)
preco = arquivo['quantidade'] * arquivo['preco_unitario']
valor_total_mes = preco.sum()
print(f"Valor vendido no mes R${valor_total_mes:.2f}")
```
